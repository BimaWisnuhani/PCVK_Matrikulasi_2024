```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from google.colab.patches import cv2_imshow
print("Mengubah tingkat kecerahan citra")
print("--------------------------------")
try:
  brightness = int(input("Masukkan nilai kecerahan : "))
except ValueError:
  print("Err0r, not a number")

original = cv.imread('/content/drive/MyDrive/PCVK/hewan.jpg')
brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat([original, brightness_image])
cv2_imshow(final_frame)
```

```python
# 2 Implementasikan transformasi contrast pada Google Colaboratory
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('/content/drive/MyDrive/PCVK/hewan.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
brightness = int(input("Masukkan tingkat kecerahan [-255 - 255]: "))
contrast = float(input("Masukkan kontras [1.0 - 3.0]: "))
def adjust_contrast_brightness(image, contrast, brightness):
    new_image = np.clip(contrast * image + brightness, 0, 255).astype(np.uint8)
    return new_image
adjusted_image = adjust_contrast_brightness(img_rgb, contrast, brightness)
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img_rgb)
plt.title('Gambar Asli')
plt.subplot(122)
plt.imshow(adjusted_image)
plt.title(f'Kontras {contrast}, Kecerahan {brightness}')
plt.show()
```

```python
# 3 Implementasikan transformasi logarithmic brightness pada Google Colaboratory
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar yang diupload
img = cv.imread('/content/drive/MyDrive/PCVK/hewan.jpg')

img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

brightness = int(input("Masukkan nilai kecerahan [0 - 100]: "))
def log_transform(image, brightness):
    image_normalized = image / 255.0
    c = 255 / np.log(1 + np.max(image_normalized))
    log_image = c * np.log(1 + image_normalized)
    log_image_bright = np.clip(log_image * 255 + brightness, 0, 255).astype(np.uint8)
    return log_image_bright

# Terapkan transformasi logaritmik
log_image = log_transform(img_rgb, brightness)
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img_rgb)
plt.title('Gambar Asli')
plt.subplot(122)
plt.imshow(log_image)
plt.title(f'Transformasi Log dengan Kecerahan {brightness}')

plt.show()
```

```python
# 4 Implementasikan transformasi grayscale menggunakan metode averaging, lightness,dan luminance pada Google Colaboratory
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('/content/drive/MyDrive/PCVK/hewan.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
def grayscale_averaging(image):
    return np.mean(image, axis=2).astype(np.uint8)
def grayscale_lightness(image):
    return ((np.max(image, axis=2) + np.min(image, axis=2)) / 2).astype(np.uint8)
def grayscale_luminance(image):
    return (0.21 * image[:,:,0] + 0.72 * image[:,:,1] + 0.07 * image[:,:,2]).astype(np.uint8)
gray_avg = grayscale_averaging(img_rgb)
gray_lightness = grayscale_lightness(img_rgb)
gray_luminance = grayscale_luminance(img_rgb)

# Tampilkan hasil
plt.figure(figsize=(15, 10))
plt.subplot(221), plt.imshow(img_rgb), plt.title('Original')
plt.subplot(222), plt.imshow(gray_avg, cmap='gray'), plt.title('Averaging')
plt.subplot(223), plt.imshow(gray_lightness, cmap='gray'), plt.title('Lightness')
plt.subplot(224), plt.imshow(gray_luminance, cmap='gray'), plt.title('Luminance')

plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('/content/drive/MyDrive/PCVK/Dongkerr.jpg')

img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

def blue_highlight(image):
    gray = (0.21 * image[:,:,0] + 0.72 * image[:,:,1] + 0.07 * image[:,:,2]).astype(np.uint8)
    blue_mask = (image[:,:,2] > 100) & (image[:,:,2] > image[:,:,1]) & (image[:,:,2] > image[:,:,0])
    result = np.dstack([gray, gray, gray])
    result[blue_mask] = image[blue_mask]
    return result
blue_highlighted = blue_highlight(img_rgb)
plt.figure(figsize=(12, 6))
plt.subplot(121), plt.imshow(img_rgb), plt.title('Original')
plt.subplot(122), plt.imshow(blue_highlighted), plt.title('Blue Highlighted')
plt.show()
```
